Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Importing dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

Using elbow method to find the optimal value of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Training the model

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_pred = kmeans.fit_predict(X)
```

```python
print(y_pred)
```

Visualizing the clusters

```python
plt.scatter(X[y_pred==0,0], X[y_pred==0,1], s = 50, c='red', label='Cluster 1')
plt.scatter(X[y_pred==1,0], X[y_pred==1,1], s = 50, c='green', label='Cluster 2')
plt.scatter(X[y_pred==2,0], X[y_pred==2,1], s = 50, c='blue', label='Cluster 3')
plt.scatter(X[y_pred==3,0], X[y_pred==3,1], s = 50, c='pink', label='Cluster 4')
plt.scatter(X[y_pred==4,0], X[y_pred==4,1], s = 50, c='cyan', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c='yellow', label='Centroids')
plt.xlabel('Annual Income (K)')
plt.ylabel('Spending Score (1-100)')
plt.title('Clusters of customers')
plt.legend()
plt.show()
```
